```python
# READ DATA 
def read_data_from_txt(file_path):
    """Reads data from a txt file with space-separated values for Tutor_names, Course_names, Timeslot_names, Preferences, and W."""
    tutor_names = []
    course_names = []
    timeslot_names = []
    preferences = {}
    w_values = {}  # Dictionary to store workshop availability (W)
    w_values = []  # List to store W values (0 or 1)

    with open(file_path, 'r') as file:
        lines = file.readlines()

        # Extract Tutor_names
        for line in lines:
            if line.startswith('Tutor_names:'):
                tutor_names_str = line.split(':')[1].strip()  # Get the list part after the colon
                tutor_names_str = tutor_names_str[1:-1]  # Remove square brackets
                tutor_names = [tutor.strip() for tutor in tutor_names_str.split("'") if tutor.strip()]

            # Extract Course_names
            if line.startswith('Course_names:'):
                course_names_str = line.split(':')[1].strip()  # Get the list part after the colon
                course_names_str = course_names_str[1:-1]  # Remove square brackets
                course_names = [course.strip() for course in course_names_str.split("'") if course.strip()]
            
            # Extract Preferences
            if line.startswith('lpt:'):
                preferences_str = line.split(':')[1].strip()  # Get the list part after the colon
                preferences_str = preferences_str[1:-1]  # Remove square brackets
                preferences_list = [float(val) for val in preferences_str.split() if val]  # Convert to float
                # Assuming preferences are per tutor, assign them accordingly
                for i, tutor in enumerate(tutor_names):
                    preferences[tutor] = preferences_list[i]  # Map each tutor to their preference value

            # Extract W values (spread across multiple lines)


            if line.startswith('W:'):
                # Skip the line with 'W:' and start reading the matrix
                continue
            elif line.strip():  # Only process non-empty lines
                # Process each value in the line, ensuring it's an integer and valid
                for val in line.strip().split():
                    try:
                        w_values.append(int(val))  # Convert each value to an integer
                    except ValueError:
                        continue  # Skip any invalid values (e.g., '!')
                





        # Define the list of timeslot names manually
    timeslot_names = [
        'Monday 9:00 - 9:30', 'Monday 9:30 - 10:00', 'Monday 10:00 - 10:30', 'Monday 10:30 - 11:00',
        'Monday 11:00 - 11:30', 'Monday 11:30 - 12:00', 'Monday 12:00 - 12:30', 'Monday 12:30 - 13:00',
        'Monday 13:00 - 13:30', 'Monday 13:30 - 14:00', 'Monday 14:00 - 14:30', 'Monday 14:30 - 15:00',
        'Monday 15:00 - 15:30', 'Monday 15:30 - 16:00', 'Monday 16:00 - 16:30', 'Monday 16:30 - 17:00',
        'Monday 17:00 - 17:30', 'Monday 17:30 - 18:00', 'Tuesday 9:00 - 9:30', 'Tuesday 9:30 - 10:00',
        'Tuesday 10:00 - 10:30', 'Tuesday 10:30 - 11:00', 'Tuesday 11:00 - 11:30', 'Tuesday 11:30 - 12:00',
        'Tuesday 12:00 - 12:30', 'Tuesday 12:30 - 13:00', 'Tuesday 13:00 - 13:30', 'Tuesday 13:30 - 14:00',
        'Tuesday 14:00 - 14:30', 'Tuesday 14:30 - 15:00', 'Tuesday 15:00 - 15:30', 'Tuesday 15:30 - 16:00',
        'Tuesday 16:00 - 16:30', 'Tuesday 16:30 - 17:00', 'Tuesday 17:00 - 17:30', 'Tuesday 17:30 - 18:00',
        'Wednesday 9:00 - 9:30', 'Wednesday 9:30 - 10:00', 'Wednesday 10:00 - 10:30', 'Wednesday 10:30 - 11:00',
        'Wednesday 11:00 - 11:30', 'Wednesday 11:30 - 12:00', 'Wednesday 12:00 - 12:30', 'Wednesday 12:30 - 13:00',
        'Wednesday 13:00 - 13:30', 'Wednesday 13:30 - 14:00', 'Wednesday 14:00 - 14:30', 'Wednesday 14:30 - 15:00',
        'Wednesday 15:00 - 15:30', 'Wednesday 15:30 - 16:00', 'Wednesday 16:00 - 16:30', 'Wednesday 16:30 - 17:00',
        'Wednesday 17:00 - 17:30', 'Wednesday 17:30 - 18:00', 'Thursday 9:00 - 9:30', 'Thursday 9:30 - 10:00',
        'Thursday 10:00 - 10:30', 'Thursday 10:30 - 11:00', 'Thursday 11:00 - 11:30', 'Thursday 11:30 - 12:00',
        'Thursday 12:00 - 12:30', 'Thursday 12:30 - 13:00', 'Thursday 13:00 - 13:30', 'Thursday 13:30 - 14:00',
        'Thursday 14:00 - 14:30', 'Thursday 14:30 - 15:00', 'Thursday 15:00 - 15:30', 'Thursday 15:30 - 16:00',
        'Thursday 16:00 - 16:30', 'Thursday 16:30 - 17:00', 'Thursday 17:00 - 17:30', 'Thursday 17:30 - 18:00',
        'Friday 9:00 - 9:30', 'Friday 9:30 - 10:00', 'Friday 10:00 - 10:30', 'Friday 10:30 - 11:00',
        'Friday 11:00 - 11:30', 'Friday 11:30 - 12:00', 'Friday 12:00 - 12:30', 'Friday 12:30 - 13:00',
        'Friday 13:00 - 13:30', 'Friday 13:30 - 14:00', 'Friday 14:00 - 14:30', 'Friday 14:30 - 15:00',
        'Friday 15:00 - 15:30', 'Friday 15:30 - 16:00', 'Friday 16:00 - 16:30', 'Friday 16:30 - 17:00',
        'Friday 17:00 - 17:30', 'Friday 17:30 - 18:00'
    ]

    # Now map W values to (course, timeslot) pairs
    w_dict = {}
    for i, w in enumerate(w_values):
        course = course_names[i % len(course_names)]  # Wrap around if more W values than courses
        timeslot = timeslot_names[i % len(timeslot_names)]  # Wrap around if more W values than timeslots
        w_dict[(course, timeslot)] = w  # Store as a tuple (course, timeslot) with corresponding W value

    max_hours = {tutor: 10 for tutor in tutor_names}  # Default value for each tutor

    return tutor_names, course_names, timeslot_names, max_hours, preferences, w_dict

# Usage example
file_path = "NewData2.txt"
tutor_names, course_names, timeslot_names, max_hours, preferences, W = read_data_from_txt(file_path)

# Printing the results
print("Tutor Names:", tutor_names)
print("Course Names:", course_names)
print("Timeslot Names:", timeslot_names)
print("Preferences:", preferences)
print("Maximum hours:", max_hours)
print("Workshop Matrix W:", W)

```

```python
# READ AVAIL AND ASSIGN OUTPUT 
import csv

def read_availability(file_path="updated_availability.csv"):
    """Reads tutor availability data from CSV and returns a dictionary."""
    availability = {}

    with open(file_path, mode='r') as file:
        reader = csv.reader(file)
        next(reader)  # Skip header row

        for row in reader:
            tutor = row[0]
            timeslot = int(row[1])  # Convert timeslot to integer
            available = int(row[2])  # Convert availability (1=Yes, 0=No) to integer

            if tutor not in availability:
                availability[tutor] = {}  # Create a sub-dictionary for the tutor
            
            availability[tutor][timeslot] = available  # Store availability for the timeslot

    return availability

def read_assignments(file_path="tutor_assignments.csv"):
    """Reads tutor assignment data from CSV and returns a dictionary."""
    assignments = {}

    with open(file_path, mode='r') as file:
        reader = csv.reader(file)
        next(reader)  # Skip header row

        for row in reader:
            tutor = row[0]
            course = row[1]
            timeslot = int(row[2])  # Convert timeslot to integer

            if tutor not in assignments:
                assignments[tutor] = []  # Create a list for the tutor
            
            assignments[tutor].append((course, timeslot))  # Store course-timeslot pair

    return assignments

# Example Usage
availability_data = read_availability()
assignments_data = read_assignments()

# Print Output
print("Availability Data:", availability_data)
print("Assignments Data:", assignments_data)
```

```python
def read_preferences(file_path):
    preferences = {}
    tutor_names = []  # List of tutors (if you need them)

    # Open the file and read line by line
    with open(file_path, 'r') as file:
        lines = file.readlines()
        
        for line in lines:
            if line.startswith('p:'):
                # Extract the preferences string and remove the 'p:' part
                preferences_str = line.split(':')[1].strip()  # Get the list part after the colon
                preferences_str = preferences_str[1:-1]  # Remove square brackets
                
                # Now split the string to get a list of preferences
                preference_rows = preferences_str.split('] [')  # Split by each row
                
                # Convert each preference row into a list of floats
                for i, row in enumerate(preference_rows):
                    # Clean up the row (remove extra spaces, brackets, etc.)
                    row = row.replace('[', '').replace(']', '').strip()
                    preferences_list = [float(val) for val in row.split()]
                    
                    # Assuming tutor_names are in the same order as the preferences
                    preferences[f'tutor_{i+1}'] = preferences_list  # Map each tutor to their preference values
                    tutor_names.append(f'tutor_{i+1}')  # Add to the tutor names list
    
    return preferences, tutor_names

# Example usage:
file_path = "NewData2.txt"
preferences, tutor_names = read_preferences(file_path)

# Print out the preferences to verify
print("Preferences:", preferences)
print("Tutor Names:", tutor_names)
```

```python
import csv
import random

def read_tutor_hours_from_csv(file_path):
    """
    Reads tutor hours from the CSV file.
    The CSV format is assumed to have two columns: tutor_name and hours_worked,
    and the first row is a header.
    
    Returns a dictionary where keys are tutor names and values are their hours worked.
    """
    tutor_hours = {}
    try:
        with open(file_path, mode='r') as file:
            csv_reader = csv.reader(file)
            next(csv_reader)  # Skip the header row
            
            for row in csv_reader:
                if row:  # Skip empty rows
                    tutor_name = row[0].strip()  # Tutor name
                    hours_worked = float(row[1].strip())  # Hours worked
                    tutor_hours[tutor_name] = hours_worked
    except FileNotFoundError:
        print(f"Error: The file {file_path} was not found.")
    return tutor_hours

def find_replacement_tutor(exited_tutor, assignments_data, availability_data, max_hours, preferences, tutor_hours_file):
    """
    Finds a replacement tutor for an exited tutor, considering tutor hours from CSV and ensuring tutors do not exceed 10 hours.
    
    - Checks for available tutors who have the capacity to take on additional slots.
    - Selects the best replacement based on the LOWEST preference score (lower is better).
    - Ensures no tutor exceeds 10 hours of work per week.
    - Ensures the same tutor replaces for all time slots of the same course.
    """
    
    # Read the tutor hours from CSV file
    tutor_hours = read_tutor_hours_from_csv(tutor_hours_file)
    
    if exited_tutor not in assignments_data:
        print(f"Tutor {exited_tutor} has no existing assignments to replace.")
        return None

    new_assignments = []
    replaced_courses = assignments_data.pop(exited_tutor)  # Remove exited tutor from assignments

    # Track the hours for each tutor (initially set to the values from the CSV)
    tutor_hours = {tutor: tutor_hours.get(tutor, 0) for tutor in assignments_data}

    # Group timeslots by course
    course_timeslots = {}
    for course, timeslot in replaced_courses:
        if course not in course_timeslots:
            course_timeslots[course] = []
        course_timeslots[course].append(timeslot)

    for course, timeslots in course_timeslots.items():
        # Find available tutors who are available for all timeslots of the course
        available_tutors = [
            t for t in availability_data if all(timeslot in availability_data[t] and availability_data[t][timeslot] == 1 for timeslot in timeslots)
        ]
        
        # Filter tutors who have not exceeded max hours (10 hours max)
        suitable_tutors = [
            t for t in available_tutors if tutor_hours.get(t, 0) + len(timeslots) * 0.5 <= 9
        ]

        if suitable_tutors:
            # Select the tutor with the **lowest preference score** for the course
            best_tutor = min(suitable_tutors, key=lambda t: preferences.get(t, {}).get(course, float("inf")))

            # Assign the new tutor to all timeslots of the course
            for timeslot in timeslots:
                assignments_data.setdefault(best_tutor, []).append((course, timeslot))
                availability_data[best_tutor][timeslot] = 0  # Update availability
                tutor_hours[best_tutor] += 0.5  # Add 0.5 hours for each timeslot

                new_assignments.append((best_tutor, course, timeslot))
                print(f"Replaced {exited_tutor} with {best_tutor} for {course} at timeslot {timeslot}")
        else:
            print(f"No available replacement found for {course} at timeslots {timeslots}")

    # Print the new hours for each tutor after replacement
    print("\nUpdated hours for each tutor:")
    for tutor, hours in tutor_hours.items():
        if hours > 10:
            print(f"{tutor} exceeded max hours. Hours: {hours}")
        else:
            print(f"{tutor}: {hours} hours")

    return new_assignments

# Example usage:
exited_tutor = input("Enter the name of the tutor who exited (or press Enter to randomize): ").strip()
if not exited_tutor:
    exited_tutor = random.choice(list(assignments_data.keys()))  # Randomly select a tutor if none provided
    print(f"No tutor entered. Randomly selected tutor: {exited_tutor}")

tutor_hours_file = 'each_tutors_hours.csv'  # Path to your CSV file

print(f"Tutor {exited_tutor} exited. Finding replacement...")

replacement_assignments = find_replacement_tutor(exited_tutor, assignments_data, availability_data, max_hours, preferences, tutor_hours_file)

if replacement_assignments:
    print("\nReplacements made:")
    for tutor, course, timeslot in replacement_assignments:
        print(f"{tutor} assigned to {course} at timeslot {timeslot}")
else:
    print("No suitable replacement found.")

```
